### Exploratory Data Analysis on Hotel Data-Set

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
file = pd.read_csv('Documents/DATA ANALYSIS PROJECTS/hotel_booking.csv')
```

```python
file.columns
```

```python
file.drop('name',inplace=True,axis=1)
file.drop('email',inplace=True,axis=1)
file.drop('phone-number',inplace=True,axis=1)
file.drop('credit_card',inplace=True,axis=1)
```

```python
file.columns
```

```python
file['reservation_status_date'] = pd.to_datetime(file.reservation_status_date)
```

```python
file.info()
```

```python
file.describe()
```

```python
file = file[file['adr']<5000]
```

```python
file.describe()
```

```python
file.info()
```

```python
PER_cancel = file.is_canceled.value_counts(normalize=True)
```

```python
PER_cancel
```

### Barchart of Canceled and Not Cancelled Rate

```python
plt.bar(['CANCELED','NOT CANCELED'],file.is_canceled.value_counts(normalize=True),edgecolor='k')
```

```python
M2015 = file[file['arrival_date_year']==2015]
M2016 = file[file['arrival_date_year']==2016]
M2017 = file[file['arrival_date_year']==2017]
```

```python
M2015.dropna()
M2016.dropna()
M2017.dropna()
```

```python
per_month_15 = M2015[M2015['is_canceled']==0]['arrival_date_month'].value_counts()
per_month_16 = M2016[M2016['is_canceled']==0]['arrival_date_month'].value_counts()
per_month_17 = M2017[M2017['is_canceled']==0]['arrival_date_month'].value_counts()
per_month_15
```

```python
per_month = file[file['is_canceled']==0]['arrival_date_month'].value_counts()
per_month
```

### CANCELATION RATE PER MONTH(OVERALL)

```python
plt.figure(figsize=(13,5))
plt.bar(file.arrival_date_month.unique(),per_month)
```

### Cancelation Rate per Month(Each Year)

```python
plt.figure(figsize=(13,5))
plt.bar(M2015.arrival_date_month.unique(),per_month_15,label='2015')
plt.bar(M2016.arrival_date_month.unique(),per_month_16,label='2016')
plt.bar(M2017.arrival_date_month.unique(),per_month_17,label = '2017')
plt.legend()
```

### Top 10 Countries With high Cancelation Rate

```python
cancel = file[file['is_canceled']==1]
top = cancel['country'].value_counts()[:10]
plt.figure(figsize=(8,8))
plt.pie(top,labels=top.index,autopct='%1.f%%')
plt.show()
```
